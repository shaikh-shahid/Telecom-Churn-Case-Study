# Telecom Churn Case Study

## Problem Statement
In the telecom industry, customers are able to choose from multiple service providers and actively switch from one operator to another. In this highly competitive market, the telecommunications industry experiences an average of 15-25% annual churn rate. Given the fact that it costs 5-10 times more to acquire a new customer than to retain an existing one, customer retention has now become even more important than customer acquisition.

For many incumbent operators, retaining high profitable customers is the number one business goal. To reduce customer churn, telecom companies need to predict which customers are at high risk of churn. In this project, you will analyze customer-level data of a leading telecom firm, build predictive models to identify customers at high risk of churn.

## Goal
It is your job to predict if a customer will churn, given the ~170 columns containing customer behavior, usage patterns, payment patterns, and other features that might be relevant. Your target variable is "churn_probability"

## 6 steps to develop prediction model
1. Data Cleaning
2. Data Analysis
3. Data Scalling
4. Model Building
5. Model Evaluation
6. Model Testing

```python
# Iterate over columns and replace NaN values with mean for numeric columns
for col in telecom_diff.columns:
    if pd.api.types.is_numeric_dtype(telecom_diff[col]):
        telecom_diff[col].fillna(telecom_diff[col].mean(), inplace=True)

# If any column is a datetime dtype, replace its NaN values with mode
for col in telecom_diff.columns:
    if pd.api.types.is_datetime64_any_dtype(telecom_diff[col]):
        mode_date = telecom_diff[col].mode()[0]
        telecom_diff[col].fillna(mode_date, inplace=True)
        
```

# Data Scaling

```python
scaler = StandardScaler()

date_columns = ['last_date_of_month_6','last_date_of_month_7', 'last_date_of_month_8',  'date_of_last_rech_6','date_of_last_rech_7', 'date_of_last_rech_8','date_of_last_rech_data_6','date_of_last_rech_data_7', 'date_of_last_rech_data_8']

# Assuming datetime_cols is a list of columns you want to transform
for col in date_columns:
    # Convert the training data to ordinal
    X_train[col] = pd.to_datetime(X_train[col])
    ordinal_dates_train = X_train[col].apply(lambda x: x.toordinal()).values.reshape(-1, 1)
    
    # Fit the scaler on training data
    scaler.fit(ordinal_dates_train)
    
    # Transform the training data
    X_train[col + '_scaled'] = scaler.transform(ordinal_dates_train)
    
    # Transform the test data
    X_test[col] = pd.to_datetime(X_test[col])
    ordinal_dates_test = X_test[col].apply(lambda x: x.toordinal()).values.reshape(-1, 1)
    X_test[col + '_scaled'] = scaler.transform(ordinal_dates_test)
```

```python
# Select only numeric columns
numeric_cols = X_train.select_dtypes(include=[np.number])

# Identify columns with zero variance in these numeric columns
columns_to_drop = numeric_cols.var()[numeric_cols.var() == 0].index.tolist()

# Drop those columns from `telecom_diff`
X_train = X_train.drop(columns=columns_to_drop)
X_test = X_test.drop(columns=columns_to_drop)
```

```python
# Drop original id and date columns

X_train = X_train.drop(columns=['last_date_of_month_6', 'id','last_date_of_month_7', 'last_date_of_month_8',  'date_of_last_rech_6','date_of_last_rech_7', 'date_of_last_rech_8','date_of_last_rech_data_6','date_of_last_rech_data_7', 'date_of_last_rech_data_8'])

X_test = X_test.drop(columns=['last_date_of_month_6', 'id','last_date_of_month_7', 'last_date_of_month_8',  'date_of_last_rech_6','date_of_last_rech_7', 'date_of_last_rech_8','date_of_last_rech_data_6','date_of_last_rech_data_7', 'date_of_last_rech_data_8'])
```

# Build PCA

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(random_state=42)
```

```python
pca.fit_transform(X_train)
```

```python
pca.explained_variance_ratio_
```

```python
pca.components_
```

```python
## scree plot
var_cumu = np.cumsum(pca.explained_variance_ratio_)
```

```python
fig = plt.figure(figsize=[12,8])
plt.plot(var_cumu)
plt.ylabel("Cummulative variance explaination")
```

```python
# Plot the cumulative explained variance
plt.plot(var_cumu, marker='o')
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("Cumulative Explained Variance vs. Number of Components")

# Find the number of components that explain a certain threshold (e.g., 95%)
threshold = 0.97
best_n_components = np.argmax(var_cumu >= threshold) + 1
plt.axvline(x=best_n_components, color='r', linestyle='--', label=f'Best n_components: {best_n_components}')

plt.legend()
plt.grid()
plt.show()

print(f"The best number of components for {threshold * 100}% explained variance: {best_n_components}")
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
classifier_rf = RandomForestClassifier(random_state=42, n_jobs=-1)
```

```python
# Set up the hyperparameter grid
param_grid = {
    'n_estimators': [10, 50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
```

```python
# Initialize GridSearchCV
grid_search = GridSearchCV(classifier_rf, param_grid, cv=5, verbose=1, n_jobs=-1)
```

```python
# Train the model
grid_search.fit(X_train, y_train)
```

```python
score_df = pd.DataFrame(grid_search.cv_results_)
score_df.head()
```

```python
score_df.nlargest(5,"mean_test_score")
```

```python
grid_search.best_estimator_
```

```python
def evaluate_model(dt_classifier):
    print("Train Accuracy :", accuracy_score(y_train, dt_classifier.predict(X_train)))
    print("Train Confusion Matrix:")
    print(confusion_matrix(y_train, dt_classifier.predict(X_train)))
    print("-"*50)
    print("Test Accuracy :", accuracy_score(y_test, dt_classifier.predict(X_test)))
    print("Test Confusion Matrix:")
    print(confusion_matrix(y_test, dt_classifier.predict(X_test)))
```

```python
rf_best = grid_search.best_estimator_
```

```python
evaluate_model(rf_best)
```

```python
# Handle datetime columns by converting to ordinal, imputing, and then converting back, if they exist
if len(datetime_cols) > 0:
    for col in datetime_cols:
        date_series = telecom_diff[col].dropna().apply(lambda x: x.toordinal())
        median_date = np.median(date_series)
        telecom_diff[col] = telecom_diff[col].apply(lambda x: median_date if pd.isna(x) else x.toordinal())
        telecom_diff[col] = pd.to_datetime(telecom_test[col], origin='unix', unit='D')
```
